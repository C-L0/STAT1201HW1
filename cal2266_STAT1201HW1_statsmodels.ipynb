```python
def plot_loghist(x, bins, title, xLabel, yLabel):
    hist, bins = np.histogram(x, bins=bins)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.title(title)
    plt.ylabel(yLabel)
    plt.xlabel(xLabel)
    plt.hist(x, bins=logbins)
    plt.xscale('log')
```

```python
plot_loghist(df3['Weight'], 12, "Cat and Dog Weight Histogram", "Log Scale", "Number of Cats and Dogs")
```

```python
def plot_hist(x, bins, title, xLabel, yLabel):
    hist, bins = np.histogram(x, bins=bins)
    plt.title(title)
    plt.ylabel(yLabel)
    plt.xlabel(xLabel)
    plt.hist(x)
    plt.xscale('linear')
```

```python
plot_hist(df3['Weight'], 12, "Cat and Dog Weight Histogram", "Linear Scale", "Number of Cats and Dogs")
```
